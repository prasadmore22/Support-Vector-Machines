```python
train1 = train.iloc[:,0:13]
train1 = pd.get_dummies(train1)
train1
```

```python
test1 = test.iloc[:,0:13]
test1 = pd.get_dummies(test1)
test1
```

```python
 # col are more, lets run PCA feature engg to decide best features
```

```python
# min max scaler
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
train_scaler = pd.DataFrame(trans.fit_transform(train1))
test_scaler = pd.DataFrame(trans.fit_transform(test1))
```

```python
pca_train = PCA(n_components = 102)
pca_train_values = pca_train.fit_transform(train_scaler)
var = pca_train.explained_variance_ratio_
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca_test = PCA(n_components = 102)
pca_test_values = pca_test.fit_transform(test_scaler)
```

```python
plt.figure(figsize=(10,5))
plt.plot(var1)
```

```python
finaltrain = pd.concat([pd.DataFrame(pca_train_values[:,0:50]),
                     train[['Salary']]], axis = 1)
finaltrain
```

```python
finaltest = pd.concat([pd.DataFrame(pca_test_values[:,0:50]),
                     test[['Salary']]], axis = 1)
finaltest
```

```python
# will use some part of data as comands are taking really long time to execute with such big data set
array = finaltrain.values
X = array[0:2000:,0:50]
Y = array[0:2000:,50]
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X,Y)
```

```python
gsv.best_params_ , gsv.best_score_
```

```python
array1 = finaltest.values
x = array1[0:2000:,0:50]
y = array1[0:2000:,50]
```

```python
clf = SVC(C= 15, gamma = 50)
clf.fit(x , y)
y_pred = clf.predict(x)
acc = accuracy_score(y, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y, y_pred)
```

```python
#Scaling the data leaving out the target variable and the taking only the numerical data for input
df1= df.iloc[:,2:30]

# Converting into numpy array
df2 = df1.values
df2
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)

df_norm = norm_func(df2)
```

```python
pca = PCA(n_components = 28)
pca_values = pca.fit_transform(df_norm)
```

```python
var = pca.explained_variance_ratio_ 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1)
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3']),
                     df[['size_category']]], axis = 1)
finalDf
```

```python
array = finalDf.values
X = array[:,0:3]
Y = array[:,3]
```

```python
test_size = 0.33
seed = 8
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)

model = LogisticRegression()
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
np.round(result, 4)
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X_train,Y_train)
```

```python
gsv.best_params_ , gsv.best_score_
```

```python
clf = SVC(C= 15, gamma = 50)
clf.fit(X_train , Y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(Y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(Y_test, y_pred)
```
